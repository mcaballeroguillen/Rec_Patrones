```python
import numpy as np
from PIL import Image
import pylab
Im_traing= ["-005E-10","-020E-10","-070E+45","+000E+00","+005E-10","+020E-10","+070E+45"]
Im_test=["-110E+65","+110E+65"]
Traing=[]
Traing_y=[]
for x in range(28):
    for s in Im_traing:
        sujeto=11+x
        if(sujeto==14):
            continue
        img_d="/home/marco/Base_Patrones/CroppedYale/"+"yaleB"+str(sujeto)+"/yaleB"+str(sujeto)+"_P00A"+s+".pgm"
        Traing.append(img_d)
        Traing_y.append(sujeto)
Test=[]
Test_y=[]
for x in range(28):
    for s in Im_test:
        sujeto=11+x
        if(sujeto==14):
            continue
        img_d="/home/marco/Base_Patrones/CroppedYale/"+"yaleB"+str(sujeto)+"/yaleB"+str(sujeto)+"_P00A"+s+".pgm"
        Test.append(img_d)
        Test_y.append(sujeto)
print(len(Traing))
print(len(Test))
```

```python
im = np.array(Image.open(Traing[0]))
m,n = im.shape[0:2]
imnbr = len(Traing)
Traing_matrix = np.array([np.array(Image.open(Traing[i])).flatten() for i in range(imnbr)])
imnbr = len(Test)
Test_matrix = np.array([np.array(Image.open(Test[i])).flatten() for i in range(imnbr)])
```

```python
Traing_matrix.shape
```

```python
from sklearn.decomposition import PCA


def myPCA(M,k):
    pca = PCA(n_components = k)
    pca.fit(M)
    return pca.components_ , pca.mean_

    
```

```python
def getProjection(I,V,means,k):
    center = I-means
    new=center
    resp=[new]
    for x in range(k):
        new=new+V[x]
        resp.append(new)
    return resp
    
```

```python
V, mean = myPCA(Traing_matrix,15)

resp = getProjection(Traing_matrix[3],V,mean,10)
print(len(resp))
```

```python
mean.shape
```

```python
immean = Traing_matrix[3].reshape(m,n)
mode = resp[10].reshape(m,n)

```

```python
pylab.figure()
pylab.gray()
pylab.imshow(immean)
pylab.figure()
pylab.gray()
pylab.imshow(mode)
pylab.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
n_neighbors = 7
 
knn = KNeighborsClassifier(n_neighbors)
knn.fit(Traing_matrix, np.array(Traing_y))

print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(Traing_matrix, np.array(Traing_y))))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(Test_matrix, np.array(Test_y))))
```

```python
pred = knn.predict(Test_matrix)
print(confusion_matrix(np.array(Test_y), pred))
print(classification_report(np.array(Test_y), pred))
```

```python
k=1
XN_train=np.matrix(np.empty(shape=((0,32256))))
XN_test = np.matrix(np.empty(shape=((0,32256))))
x=0
for im in Traing_matrix:
    V, mean = myPCA(Traing_matrix,k)
    resp = getProjection(im,V,mean,k)
    pro=resp[k-1]
    XN_train= np.insert(XN_train,x,pro,axis=0)
    x=x+1

x=0
for im in Traing_matrix:
    V, mean = myPCA(Traing_matrix,k)
    resp = getProjection(im,V,mean,k)
    pro=resp[k-1]
    XN_train= np.insert(XN_train,x,pro,axis=0)
    x=x+1
```
